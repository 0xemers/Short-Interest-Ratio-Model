## Cap-Weighted Portfolio

```python
DATA = pd.read_excel('DATA.xlsx')
DATA = DATA.sort_values(by=['DT','STOCK_NAME']).reset_index().drop(columns = 'index')
DATA.drop(DATA.head(504*46).index,inplace=True)
DATA = DATA.reset_index().drop(columns = 'index')
DATA = DATA.rename(columns={'DT':'DATE'})
DATA
```

```python
CWP['SUBS_1M_USD'] = DATA['SUBS_1M_USD']
CWP
```

```python
# Monthly performance of Cap-Weighted Portfolio
def weight_avg(x, weights):
    weights[x.isnull()==True]=0
    x = x.fillna(0)
    return sum(x*weights)/sum(weights)
CWP_perf = CWP.groupby(['DATE'])[['SUBS_1M_USD','Wgt_Port']].apply(lambda x: weight_avg(x['SUBS_1M_USD'],weights=x['Wgt_Port'])).to_frame()
CWP_perf.columns = ['perf_Port']
CWP_perf['perf_BM'] = CWP.groupby(['DATE'])[['SUBS_1M_USD','Wgt_BM']].apply(lambda x: weight_avg(x['SUBS_1M_USD'],weights=x['Wgt_BM'])).to_frame()
CWP_perf['perf_Rel'] = CWP_perf['perf_Port'] - CWP_perf['perf_BM']
CWP_perf.head()
```

```python
CWP_Annual_return = CWP_perf['perf_Port'].mean()*12
CWP_Annual_volatility = (CWP_perf['perf_Port'].var()*12)**0.5
CWP_ExRtn = CWP_perf['perf_Rel'].mean()*12
CWP_RiskAdjRtn = CWP_ExRtn/(CWP_perf['perf_Rel'].std()*12**0.5)
```

# Historical risk profile

```python
# Factor model
Rtn_fac = pd.read_excel("Rtn_fac.xlsx")
Rtn_fac=Rtn_fac.drop(columns='Unnamed: 0')
Rtn_fac=Rtn_fac.drop(columns='Low')
Rtn_fac.head()
```

```python
# Orthogonalization
def regress(X,Y):
    X_inv=pd.DataFrame(np.linalg.pinv(X.values),X.columns,X.index)
    B= X_inv.dot(Y)
    resids=Y-X.dot(B)
    factors=pd.concat([X,resids],axis=1)
    return factors,B

MKT=['SPX Index', 'VIX Index', 'SPX Index Volume']
SCT=['Energy', 'Materials', 'Industrials', 'Consumer Discretionary', 'Consumer Staples', 'Health Care', 'Financials','Information Technology', 'Telecommunication Services', 'Utilities','Real Estate']
MACRO=['USGG10YR Index','USGG12M Index', 'USO US Equity', 'CONSSENT Index']
STY=['Short Interest']

def orthognal(Rtn_fac):
    fac_orth,Beta1=regress(pd.DataFrame({'intercept':1,'SPX Index':Rtn_fac['SPX Index'],'VIX Index':Rtn_fac['VIX Index'],'SPX Index Volume':Rtn_fac['SPX Index Volume']}),Rtn_fac[SCT])
    fac_orth,Beta2=regress(fac_orth,Rtn_fac[STY])
    fac_orth,Beta3=regress(fac_orth,Rtn_fac[MACRO])
    fac_orth=fac_orth.drop(labels='intercept',axis=1)
    return fac_orth

fac_orth=orthognal(Rtn_fac)
```

```python
# Prepare the stock return data
last_p=pd.read_excel("Data_original.xlsx",sheet_name='SPX_Last Price')
for i in range(last_p.shape[0]-1):
    last_p.iloc[i,1:] = (last_p.iloc[i+1,1:]-last_p.iloc[i,1:] )/last_p.iloc[i,1:]
last_p=last_p.drop(last_p.index[-1])
last_p=last_p.set_index(["DATES"])
last_p.columns=pd.Index(last_p.columns,name="STOCK_NAME")
```

```python
# Impute the missing data with sector return
Return=last_p.stack(dropna=False).to_frame().reset_index()
Return.columns=['DT','NAME','Return']
Return.groupby('DT').count().head()
```

```python
# Fill the Na return with sector return
sector=pd.read_excel("Data_original.xlsx",sheet_name='SPX_GICS')
sector=sector[['STOCK_NAME','Sector_name']]
sector.columns=['NAME','Sct']
sector.head()
```

```python
Return=Return.merge(sector,how='left',left_on='NAME',right_on='NAME')
rtn_sct=Rtn_fac[['DT']+SCT]
rtn_sct=pd.melt(rtn_sct,id_vars='DT')
rtn_sct.columns=['DT','Sct','rtn_sct']
Return=Return.merge(rtn_sct,how='left',left_on=['DT','Sct'],right_on=['DT','Sct'])

import math
Return['Return']=[y if math.isnan(x) else x for x,y in zip(Return.Return,Return.rtn_sct)]
Return.head()
```

```python
Return=Return.pivot(index='DT',columns='NAME',values='Return')
fac_orth.index=Return.index
```

## Cap-Weighted Portfolio

```python
window = 40
start = 6

DATE = list(CWP['DATE'].unique())
tv_sys = []
tv_spe = []
tr = []
tv_MKT = []
tv_SCT = []
tv_STY = []
tv_MACRO = []

actv_sys = []
actv_spe = []
te = []
actv_MKT = []
actv_SCT = []
actv_STY = []
actv_MACRO = []

Max_TV_fac = []
Max_ActV_fac = []

CWP_risk = pd.DataFrame(columns=['DATE','TV_sys','TV_spe','TR','TV_MKT','TV_SCT','TV_STY','TV_MACRO','ActV_sys','ActV_spe','TE','ActV_MKT','ActV_SCT','ActV_STY','ActV_MACRO'])
```

```python
for i in CWP['DATE'].unique():
    resid,beta=regress(fac_orth.iloc[start:start+window], Return[start:start+window])
    
    K=1+len(SCT)+len(STY)+len(MACRO)
    var_ido=resid.drop(labels=MKT+SCT+STY+MACRO,axis=1).apply(lambda x:sum(x**2)/(60-K),axis=0)
    var_ido=pd.DataFrame(np.diag(var_ido),var_ido.index,var_ido.index)
    
    weight = CWP.iloc[504*(start-6):504*(start-5)].drop('DATE',1).set_index('NAME')
    u=beta.dot(weight)
    
    # Total risk attribution
    F=fac_orth.cov()
    TV_sys = weight.Wgt_Port.dot((beta.T)).dot(F).dot(beta).dot(weight.Wgt_Port)*12
    TV_spe = weight.Wgt_Port.dot((var_ido)).dot(weight.Wgt_Port)*12
    TR = np.sqrt(TV_sys+TV_spe)
    
    tv_sys.append(weight.Wgt_Port.dot((beta.T)).dot(F).dot(beta).dot(weight.Wgt_Port)*12)
    tv_spe.append(weight.Wgt_Port.dot((var_ido)).dot(weight.Wgt_Port)*12)
    tr.append(np.sqrt(TV_sys+TV_spe))
    
    F_MKT=F.loc[np.isin(F.index.values,MKT),MKT]
    beta_MKT=beta.loc[np.isin(beta.index.values,MKT),:]
    TV_MKT=weight.Wgt_Port.dot(beta_MKT.T).dot(F_MKT).dot(beta_MKT).dot(weight.Wgt_Port)*12
    tv_MKT.append(weight.Wgt_Port.dot(beta_MKT.T).dot(F_MKT).dot(beta_MKT).dot(weight.Wgt_Port)*12)
    
    F_SCT=F.loc[np.isin(F.index.values,SCT),SCT]
    beta_SCT=beta.loc[np.isin(beta.index.values,SCT),:]
    TV_SCT=weight.Wgt_Port.dot(beta_SCT.T).dot(F_SCT).dot(beta_SCT).dot(weight.Wgt_Port)*12
    tv_SCT.append(weight.Wgt_Port.dot(beta_SCT.T).dot(F_SCT).dot(beta_SCT).dot(weight.Wgt_Port)*12)
    
    F_STY=F.loc[np.isin(F.index.values,STY),STY]
    beta_STY=beta.loc[np.isin(beta.index.values,STY),:]
    TV_STY=weight.Wgt_Port.dot(beta_STY.T).dot(F_STY).dot(beta_STY).dot(weight.Wgt_Port)*12
    tv_STY.append(weight.Wgt_Port.dot(beta_STY.T).dot(F_STY).dot(beta_STY).dot(weight.Wgt_Port)*12)
    
    F_MACRO=F.loc[np.isin(F.index.values,MACRO),MACRO]
    beta_MACRO=beta.loc[np.isin(beta.index.values,MACRO),:]
    TV_MACRO=weight.Wgt_Port.dot(beta_MACRO.T).dot(F_MACRO).dot(beta_MACRO).dot(weight.Wgt_Port)*12
    tv_MACRO.append(weight.Wgt_Port.dot(beta_MACRO.T).dot(F_MACRO).dot(beta_MACRO).dot(weight.Wgt_Port)*12)
    
    if start == 7:
        TV_Breakdown = TV_i.to_frame().T
        
    TV_i=[]
    for i in range(F.shape[1]):
        F_i=pd.DataFrame(0,F.columns,F.index)
        F_i.iloc[:,i]=F.iloc[:,i]
        TV_i=TV_i+[weight.Wgt_Port.dot(beta.T).dot(F_i).dot(beta).dot(weight.Wgt_Port)*12]
    TV_i=pd.Series(TV_i,F.index)
    max_TV_fac = TV_i.idxmax()
    Max_TV_fac.append(max_TV_fac)
    
    if start > 6:
        TV_Breakdown = pd.concat([TV_Breakdown,TV_i.to_frame().T])
    
    V=(beta.T.dot(F).dot(beta)+var_ido)*12
    
    if start == 7:
        sum_wMCTR = wMCTR
    MCTR=V.dot(weight.Wgt_Port)/TR
    wMCTR = MCTR*weight.Wgt_Port
    if start > 6:
        sum_wMCTR = wMCTR + sum_wMCTR
    
    # Tracking error attribution
    ActV_sys = weight.Wgt_Act.dot(beta.T).dot(F).dot(beta).dot(weight.Wgt_Act)*12
    ActV_spe = weight.Wgt_Act.dot((var_ido)).dot(weight.Wgt_Act)*12
    TE = np.sqrt(ActV_sys+ActV_spe)
    
    actv_sys.append(weight.Wgt_Act.dot(beta.T).dot(F).dot(beta).dot(weight.Wgt_Act)*12)
    actv_spe.append(weight.Wgt_Act.dot((var_ido)).dot(weight.Wgt_Act)*12)
    te.append(np.sqrt(ActV_sys+ActV_spe))
    
    ActV_MKT=weight.Wgt_Act.dot(beta_MKT.T).dot(F_MKT).dot(beta_MKT).dot(weight.Wgt_Act)*12
    ActV_SCT=weight.Wgt_Act.dot(beta_SCT.T).dot(F_SCT).dot(beta_SCT).dot(weight.Wgt_Act)*12
    ActV_STY=weight.Wgt_Act.dot(beta_STY.T).dot(F_STY).dot(beta_STY).dot(weight.Wgt_Act)*12
    ActV_MACRO=weight.Wgt_Act.dot(beta_MACRO.T).dot(F_MACRO).dot(beta_MACRO).dot(weight.Wgt_Act)*12
    actv_MKT.append(weight.Wgt_Act.dot(beta_MKT.T).dot(F_MKT).dot(beta_MKT).dot(weight.Wgt_Act)*12)
    actv_SCT.append(weight.Wgt_Act.dot(beta_SCT.T).dot(F_SCT).dot(beta_SCT).dot(weight.Wgt_Act)*12)
    actv_STY.append(weight.Wgt_Act.dot(beta_STY.T).dot(F_STY).dot(beta_STY).dot(weight.Wgt_Act)*12)
    actv_MACRO.append(weight.Wgt_Act.dot(beta_MACRO.T).dot(F_MACRO).dot(beta_MACRO).dot(weight.Wgt_Act)*12)
    
    ActV_i=[]
    for i in range(F.shape[1]):
        F_i=pd.DataFrame(0,F.columns,F.index)
        F_i.iloc[:,i]=F.iloc[:,i]
        ActV_i=ActV_i+[weight.Wgt_Act.dot(beta.T).dot(F_i).dot(beta).dot(weight.Wgt_Act)*12]
    ActV_i=pd.Series(ActV_i,F.index)
    max_ActV_fac = ActV_i.idxmax()
    Max_ActV_fac.append(max_ActV_fac)
    
    if start ==7:
        sum_wMCAR = wMCAR
    MCAR=V.dot(weight.Wgt_Act)/TE
    wMCAR = MCAR*weight.Wgt_Act
    if start > 6:
        sum_wMCAR = wMCAR + sum_wMCAR

    start = start + 1
```

```python
for i in MTE['DATE'].unique():
    resid,beta=regress(fac_orth.iloc[start:start+window], Return[start:start+window])
    
    K=1+len(SCT)+len(STY)+len(MACRO)
    var_ido=resid.drop(labels=MKT+SCT+STY+MACRO,axis=1).apply(lambda x:sum(x**2)/(60-K),axis=0)
    var_ido=pd.DataFrame(np.diag(var_ido),var_ido.index,var_ido.index)
    
    weight = MTE.iloc[504*(start-6):504*(start-5)].drop('DATE',1).set_index('NAME')
    u=beta.dot(weight)
    
    # Total risk attribution
    F=fac_orth.cov()
    TV_sys = weight.Wgt_Port.dot((beta.T)).dot(F).dot(beta).dot(weight.Wgt_Port)*12
    TV_spe = weight.Wgt_Port.dot((var_ido)).dot(weight.Wgt_Port)*12
    TR = np.sqrt(TV_sys+TV_spe)
    
    tv_sys.append(weight.Wgt_Port.dot((beta.T)).dot(F).dot(beta).dot(weight.Wgt_Port)*12)
    tv_spe.append(weight.Wgt_Port.dot((var_ido)).dot(weight.Wgt_Port)*12)
    tr.append(np.sqrt(TV_sys+TV_spe))
    
    F_MKT=F.loc[np.isin(F.index.values,MKT),MKT]
    beta_MKT=beta.loc[np.isin(beta.index.values,MKT),:]
    TV_MKT=weight.Wgt_Port.dot(beta_MKT.T).dot(F_MKT).dot(beta_MKT).dot(weight.Wgt_Port)*12
    tv_MKT.append(weight.Wgt_Port.dot(beta_MKT.T).dot(F_MKT).dot(beta_MKT).dot(weight.Wgt_Port)*12)
    
    F_SCT=F.loc[np.isin(F.index.values,SCT),SCT]
    beta_SCT=beta.loc[np.isin(beta.index.values,SCT),:]
    TV_SCT=weight.Wgt_Port.dot(beta_SCT.T).dot(F_SCT).dot(beta_SCT).dot(weight.Wgt_Port)*12
    tv_SCT.append(weight.Wgt_Port.dot(beta_SCT.T).dot(F_SCT).dot(beta_SCT).dot(weight.Wgt_Port)*12)
    
    F_STY=F.loc[np.isin(F.index.values,STY),STY]
    beta_STY=beta.loc[np.isin(beta.index.values,STY),:]
    TV_STY=weight.Wgt_Port.dot(beta_STY.T).dot(F_STY).dot(beta_STY).dot(weight.Wgt_Port)*12
    tv_STY.append(weight.Wgt_Port.dot(beta_STY.T).dot(F_STY).dot(beta_STY).dot(weight.Wgt_Port)*12)
    
    F_MACRO=F.loc[np.isin(F.index.values,MACRO),MACRO]
    beta_MACRO=beta.loc[np.isin(beta.index.values,MACRO),:]
    TV_MACRO=weight.Wgt_Port.dot(beta_MACRO.T).dot(F_MACRO).dot(beta_MACRO).dot(weight.Wgt_Port)*12
    tv_MACRO.append(weight.Wgt_Port.dot(beta_MACRO.T).dot(F_MACRO).dot(beta_MACRO).dot(weight.Wgt_Port)*12)
    
    if start == 7:
        TV_Breakdown = TV_i.to_frame().T
    TV_i=[]
    for i in range(F.shape[1]):
        F_i=pd.DataFrame(0,F.columns,F.index)
        F_i.iloc[:,i]=F.iloc[:,i]
        TV_i=TV_i+[weight.Wgt_Port.dot(beta.T).dot(F_i).dot(beta).dot(weight.Wgt_Port)*12]
    TV_i=pd.Series(TV_i,F.index)
    if start > 6:
        TV_Breakdown = pd.concat([TV_Breakdown,TV_i.to_frame().T])
    
    V=(beta.T.dot(F).dot(beta)+var_ido)*12
    
    if start ==7:
        sum_wMCTR = wMCTR
    MCTR=V.dot(weight.Wgt_Port)/TR
    wMCTR = MCTR*weight.Wgt_Port
    if start > 6:
        sum_wMCTR = wMCTR + sum_wMCTR
    
    # Tracking error attribution
    ActV_sys = weight.Wgt_Act.dot(beta.T).dot(F).dot(beta).dot(weight.Wgt_Act)*12
    ActV_spe = weight.Wgt_Act.dot((var_ido)).dot(weight.Wgt_Act)*12
    TE = np.sqrt(ActV_sys+ActV_spe)
    
    actv_sys.append(weight.Wgt_Act.dot(beta.T).dot(F).dot(beta).dot(weight.Wgt_Act)*12)
    actv_spe.append(weight.Wgt_Act.dot((var_ido)).dot(weight.Wgt_Act)*12)
    te.append(np.sqrt(ActV_sys+ActV_spe))
    
    ActV_MKT=weight.Wgt_Act.dot(beta_MKT.T).dot(F_MKT).dot(beta_MKT).dot(weight.Wgt_Act)*12
    ActV_SCT=weight.Wgt_Act.dot(beta_SCT.T).dot(F_SCT).dot(beta_SCT).dot(weight.Wgt_Act)*12
    ActV_STY=weight.Wgt_Act.dot(beta_STY.T).dot(F_STY).dot(beta_STY).dot(weight.Wgt_Act)*12
    ActV_MACRO=weight.Wgt_Act.dot(beta_MACRO.T).dot(F_MACRO).dot(beta_MACRO).dot(weight.Wgt_Act)*12
    actv_MKT.append(weight.Wgt_Act.dot(beta_MKT.T).dot(F_MKT).dot(beta_MKT).dot(weight.Wgt_Act)*12)
    actv_SCT.append(weight.Wgt_Act.dot(beta_SCT.T).dot(F_SCT).dot(beta_SCT).dot(weight.Wgt_Act)*12)
    actv_STY.append(weight.Wgt_Act.dot(beta_STY.T).dot(F_STY).dot(beta_STY).dot(weight.Wgt_Act)*12)
    actv_MACRO.append(weight.Wgt_Act.dot(beta_MACRO.T).dot(F_MACRO).dot(beta_MACRO).dot(weight.Wgt_Act)*12)
    
    ActV_i=[]
    for i in range(F.shape[1]):
        F_i=pd.DataFrame(0,F.columns,F.index)
        F_i.iloc[:,i]=F.iloc[:,i]
        ActV_i=ActV_i+[weight.Wgt_Act.dot(beta.T).dot(F_i).dot(beta).dot(weight.Wgt_Act)*12]
    ActV_i=pd.Series(ActV_i,F.index)
    
    if start == 7:
        sum_wMCAR = wMCAR
    MCAR=V.dot(weight.Wgt_Act)/TE
    wMCAR = MCAR*weight.Wgt_Act
    if start > 6:
        sum_wMCAR = wMCAR + sum_wMCAR
        
    start = start + 1
```
